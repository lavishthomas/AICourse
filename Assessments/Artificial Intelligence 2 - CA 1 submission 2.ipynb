### Q3: Regular Expression

Regular expression is representation style for sequence of characters which is used for pattern matching in a text based searching. It allows the programmers to search for patterns like phone number, area code etc.


```python
## For Eg.
## searching for dates avaialble. search methods returns the position of the first match

search_pattern = r'\d\d:\d\d'
time = re.search(search_pattern, entire_text)
print('The first time : ' + str(time))

## The findall funcctions returns al the matches in the text

search_pattern = r'\d\d:\d\d'
time = re.findall(search_pattern, entire_text)
time
```

### Answer

```python
search_patterns = [" [aA]\w+", " [eE]\w+", " [iI]\w+" ," [oO]\w+", " [uU]\w+"] 
total = 0 
for pattern in search_patterns:
    occurance = re.findall(pattern, entire_text)
    print ('The character' , pattern , 'occurred ' , len(occurance) , 'times')
    total += len(occurance)

print('\nIn total there are ' , total, 'words startng with Vowels')    
```

### Q7: Rule based matching


```python
# Import the Matcher library
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
token_match1 = [{"LOWER": "man"}]
token_match2 = [{"LOWER": "Man"}]
token_match3 = [{"LOWER": "MAN"}]
token_match4 = [{"LOWER": "MEN"}]

matcher.add("man", None, token_match1, token_match2, token_match3, token_match4 )

token_match4 = [{"LOWER": "woman"}]
token_match5 = [{"LOWER": "Woman"}]
token_match6 = [{"LOWER": "WOMAN"}]
token_match7 = [{"LOWER": "WOMEN"}]

matcher.add("woman", None, token_match4, token_match5, token_match6, token_match6 )
```

```python
### This fucntion takes the text file and matcher object as input 
### and checks for matchs int eh text file
def find_matches(text, matcher):
    pd = [];
    # find all matches within the doc object
    token_matches = matcher(text)
    # For each item in the token_matches provide the following
    # match_id is the hash value of the identified token match
    print(f"{'Match ID':<{20}} {'String':<{10}} {'Start':{5}} {'End':{5}} {'Match word':{20}}")
    for match_id, start, end in token_matches:
        string_id = nlp.vocab.strings[match_id]
        # the pattern is counted
        pd.append(string_id)
        matched_span = doc_object[start:end]      
        print(f"{match_id:<{20}} {string_id:<{10}} {start:<{5}} {end:<{5}} {matched_span.text:{20}}")
    return pd
```

```python
### find_matches and print details
counter = find_matches(doc_object, matcher)
```

### Answer

```python
print ("The word man has occured :" + str(counter.count('man')))
print ("The word woman has occured :" + str(counter.count('woman')))
```

### Q8: Phrase matching

```python
# Import the PhraseMatcher library
from spacy.matcher import PhraseMatcher
phrase_matcher = PhraseMatcher(nlp.vocab)
```

```python
phrase_list = ["laugh", "cry", "happy", "sad"]
phrase_patterns = [nlp.make_doc(word) for word in phrase_list]
phrase_matcher.add("emotion", None, *phrase_patterns)
```

```python
### This fucntion takes the text file and matcher object as input 
### and checks for matchs int eh text file

def prev_next_words(text, phrase_matcher):
    # find all matches within the doc object
    phrase_token_matches = phrase_matcher(text)
    # For each item in the token_matches provide the following
    # match_id is the hash value of the identified token match
    print(f"{'Start':{5}} {'End':{5}} {'Match word':{15}} {'Previous 5 words':{40}} {'Next 5 words':{15}}")
    for match_id, start, end in phrase_token_matches:
        ### Extarcting
        string_id = nlp.vocab.strings[match_id] 
        matched_span = doc_object[start:end]      
        ### Printing
        print(f"{start:{5}} {end:{5}} {matched_span.text:{15}} {str(doc_object[start-5:start]):{40}} {str(doc_object[end:end+5]):{15}}")
```

### Q9: Lemmatisation
Lemmatization converts words in the second or third forms to their first form variants. It applies a morphological analysis to words to convert the words into more base form. Unlike in the stemming, the root words or lemma is only generated if and only if the word is present in the dictionary of the module used.<br>

In stemming, the words are trimmed using common prefixes and suffix presnet like 'ing', 'ed' , etc. But in lemmatisation the words are converted into more base form using the dictionary method like 'good', 'better' and 'best' all could be converted to 'good'using lemmatisation.
